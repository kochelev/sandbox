```python
import psutil
import time
from IPython.display import clear_output
from prettytable import PrettyTable
from prettytable.colortable import ColorTable, Themes

max_rows = 10
cpu_count = psutil.cpu_count()
field_names_cpu_list = ['CPU %d' % x for x in range(1, cpu_count+1)]

table = ColorTable(theme=Themes.OCEAN)
table.field_names = ['#', 'Memory', 'Disk', 'CPUs', ' : '] + field_names_cpu_list

counter = 1

while True:
    
    memory_usage = psutil.virtual_memory().percent
    disk_usage = psutil.disk_usage('/').percent
    cpu_usage = psutil.cpu_percent(interval=1)
    each_cpu_usage = psutil.cpu_percent(percpu=True)

    table.add_row([counter, f'{memory_usage}%', f'{disk_usage}%', f'{cpu_usage}%', ' : ', *each_cpu_usage])

    if counter > max_rows:
        table.del_row(0)

    print(table)

    counter += 1
    
    time.sleep(1)

    clear_output(wait=True)
```
