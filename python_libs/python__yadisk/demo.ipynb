```python
import os
from dotenv import load_dotenv

load_dotenv()

token = os.getenv("YADISK_TOKEN")
token
```

```python
import yadisk

client = yadisk.YaDisk(token=token)

# client = yadisk.YaDisk("<id-приложения>", "<secret-приложения>", "<токен>")
```

### Проверка валидности токена

```python
print(client.check_token())
```

### Получение общей информации о диске

```python
info = client.get_disk_info()

print(info)
```

### Создание папки

```python
print(client.mkdir('/example_folder'))
```

### Вывод содержимого папки

```python
folders = client.listdir('/work_data')

folders_list = []
for f in folders:
    print(f)
    folders_list.append(f)
```

### Имя последнего файла

```python
folders_list[0].name
```

### Загрузка файла в папку

```python
client.upload('image.png', '/example_folder/image.png')

# with open('image.png', 'rb') as f:
#     y.upload(f, '/Загрузки/image.png')
```

### Скачивание файла

```python
client.download("/example_folder/image.png", "downloaded_image.png")
```

### Удаление файла

```python
client.remove("/example_folder/image.png", permanently=True)
```

### Удаление папки

```python
client.remove("/example_folder", permanently=True)
```
