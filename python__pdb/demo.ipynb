# Работа со встроенным отладчиком **PDB**

Команды навигации или контроля исполнения кода:

- **n** - (next) выполнить текущую строку и остановиться на следующей строке внутри текущей функции (не заходя во вложенные функции).
- **s** - (step) выполнить текущую строки и либо остановиться на следующей строке, либо войти внутрь вложенной функции.
- **c** - (continue) продолжить выполнение команд программы до следующей точки останова или завершения программы.
- **r** - (return) продолжить выполнение команд программы до выхода из текущей функции.
- **l** - (list) отобразить код вокруг текущей строки.
- **b** - (break) установить точку останова, используется с номером строки (b 10), названием функции (b my_func), или названием файла и номером строки (b my_file.py:15).
- **cl** - (clear) удалить точку останова (cl 2) или удалить все точки останова (cl).
- **unt** - (until) продолжить выполнение до строки, с номером большим, чем номер текущей строки (unt 10).
- **q** - (quit) завершить отладку и прервать выполнение программы.

Инспектирование переменных и состояний:

- **p {expression}** - (print) вывести в консоль значение выражения или переменной.
- **pp {expression}** - (pretty-print) то же, но с большей читаемостью для сложных объектов.
- **w** - (where) отобразить call stack - показать последовательность вызовов функций от начала до текущей строки.
- **a** - (args) отобразить аргументы текущей функции.
- **d** - (down) перейти на один уровень ниже по call stack.
- **u** - (up) перейти на один уровень выше по call stack.

Помощь и другое:
- **h** - (help) отобразить список всех доступных команд отладчика или помощь по конкретной команде (help next).
- **alias {name} {command} {command args}** - создать синоним для команды, включая ее аргументы.
- **unalias {name}** - удалить синоним по его названию.
- **run / restart** - перезапустить отладчик (команды привели к ошибке).

```python
import pdb

def my_func(x):
    z = x ** 2
    return z

def wrapper(var_1: int, var_2: int) -> int:
    pdb.set_trace()
    var_1 = my_func(var_1)
    var_2 = my_func(var_2)
    var_sum = var_1 + var_2
    return var_sum

wrapper(4, 3)
```
