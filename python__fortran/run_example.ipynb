# Запуск кода Fortran

Для Apple Silicon:

```shell
brew install gcc
```

Если **Python** бы установлен через **Homebrew**, то необходимо также установить **pkg-config**

```shell
brew install pkgconf
```

Далее необходимо развернуть виртуальное окружение и выполнить установку необходимых модулей:

```shell
pip install ninja meson numpy
```

### Способ 1: Прямая компиляция

```shell
f2py -c -m example example.f90
```

### Способ 2: С указанием компилятора

```shell
f2py -c -m example example.f90 --fcompiler=gfortran
```

### Способ 3: Для отладки

```shell
f2py -c -m example example.f90 --fcompiler=gfortran --opt='-O0' --debug
```
